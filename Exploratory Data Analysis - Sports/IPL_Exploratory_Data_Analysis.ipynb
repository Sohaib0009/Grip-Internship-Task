```python
ipl["season"] = pd.DatetimeIndex(ipl["date"]).year
ipl.head()
```

```python
match_per_seas=ipl.groupby(["season"])["id"].count().reset_index().rename(columns={'id':'matches'})
match_per_seas
```

```python
sns.countplot(ipl["season"],palette="mako")

plt.xticks(rotation=45,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("SEASON",fontsize=10,fontweight="bold")
plt.ylabel("MATCH",fontsize=10,fontweight="bold")
plt.title("MATCH PLAYED IN SEASONS",fontsize=10,fontweight="bold")
plt.show()
```

```python
season_data=ipl[['id','season']].merge(ball,on="id",how="right").drop('id',axis=1)
season_data.head()
```

```python
season_run=season_data.groupby(['season'])["total_runs"].sum().reset_index()
```

```python
season_run
```

```python
sr=sns.lineplot(x="season",y="total_runs",color='r',data=season_run,marker="o",markerfacecolor="k")
plt.title("RUNS PER SEASON",fontweight="bold")
plt.xlabel("SEASON",fontweight="bold")
plt.ylabel("RUNS",fontweight="bold")
plt.grid()

plt.show()
```

```python
run_per_season=match_per_seas.merge(season_run,on='season')
run_per_season['run_per_match']=run_per_season["total_runs"]/run_per_season["matches"]
run_per_season
```
